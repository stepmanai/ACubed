```python
# Verify that all tables are dropped
print("Verifying table state...")
print("="*60)

# All FFR tables across all layers
bronze_tables = [
    "acubed.ffr.bronze__songlist",
    "acubed.ffr.bronze__charts",
    "acubed.ffr.bronze__playlist",
    "acubed.ffr.bronze__contested-difficulties"
]

silver_tables = [
    "acubed.ffr.silver__playlist",
    "acubed.ffr.silver__notes",
    "acubed.ffr.silver__notes-adjusted",
    "acubed.ffr.silver__vertical-density",
    "acubed.ffr.silver__horizontal-density",
    "acubed.ffr.silver__chart-features",
    "acubed.ffr.silver__song-features",  # Old table name before rename
    "acubed.ffr.silver__reliability-scores",
    "acubed.ffr.silver__contested-difficulties"
]

gold_tables = [
    "acubed.ffr.gold__features"
]

all_tables = bronze_tables + silver_tables + gold_tables

# Helper function to check if table exists (handles hyphens in names)
def table_exists(table_name):
    try:
        spark.sql(f"DESCRIBE TABLE `{table_name}`")
        return True
    except:
        return False

# Check each layer
print("\nBRONZE LAYER:")
for table in bronze_tables:
    exists = table_exists(table)
    status = "❌ EXISTS" if exists else "✓ DROPPED"
    print(f"  {status}: {table}")

print("\nSILVER LAYER:")
for table in silver_tables:
    exists = table_exists(table)
    status = "❌ EXISTS" if exists else "✓ DROPPED"
    print(f"  {status}: {table}")

print("\nGOLD LAYER:")
for table in gold_tables:
    exists = table_exists(table)
    status = "❌ EXISTS" if exists else "✓ DROPPED"
    print(f"  {status}: {table}")

# Overall status
all_dropped = all(not table_exists(table) for table in all_tables)

print("\n" + "="*60)
if all_dropped:
    print("✅ SUCCESS: All FFR tables have been dropped")
    print("\nCatalog state: CLEAN")
    print(f"Tables dropped: {len(all_tables)}")
    print("\n" + "="*60)
else:
    print("⚠️ WARNING: Some tables still exist")
    print("\nTables still present:")
    for table in all_tables:
        if table_exists(table):
            print(f"  • {table}")
    print("\nRecommendation: Re-run the drop cells above")
```
